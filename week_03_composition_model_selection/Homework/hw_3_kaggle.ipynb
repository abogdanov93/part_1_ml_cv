````python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import make_pipeline
````

1.   train/test

````python
X = data.drop(columns=['Churn'])
y = data['Churn']

#stratify = y ..  ,          .
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# 1)    (fit  train, transform  test)
scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[num_cols])
X_test_num = scaler.transform(X_test[num_cols])

# 2)    (fit  train, transform  test)
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_train_cat = encoder.fit_transform(X_train[cat_cols])
X_test_cat = encoder.transform(X_test[cat_cols])

# 3)     
X_train_prepared = np.hstack([X_train_num, X_train_cat])
X_test_prepared = np.hstack([X_test_num, X_test_cat])

# 4)  LogisticRegressionCV
lr_cv = LogisticRegressionCV(
    Cs=[100, 10, 1, 0.1, 0.01, 0.001],
    cv=5,
    scoring='roc_auc',
    refit=True,
    max_iter=1000
)

lr_cv.fit(X_train_prepared, y_train)

print(" C:", lr_cv.C_)

# 5)     
y_test_proba = lr_cv.predict_proba(X_test_prepared)[:, 1]
roc_auc = roc_auc_score(y_test, y_test_proba)
print("ROC-AUC  :", roc_auc)
````

````python
from sklearn.compose import make_column_transformer


X = data.drop(columns=['Churn'])
y = data['Churn']

# 1)              .
preprocessor = make_column_transformer(
    (StandardScaler(), num_cols),
    (OneHotEncoder(sparse_output=False, handle_unknown='ignore'), cat_cols)
)

# 2)  ,   
pipeline = make_pipeline(
    preprocessor,
    LogisticRegression()
)

# 3)    ,
# logisticregression -     ,          (C)
param_grid = {
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001]
}

# 4) GridSearchCV:
grid_search = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=5,
    refit=True
)

# 5)  
grid_search.fit(X, y)

print("  C:", grid_search.best_params_)
print(" ROC-AUC:", grid_search.best_score_)

#     C
best_model = grid_search.best_estimator_
````

````python
X = data.drop(columns=['Churn'])
y = data['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, stratify=y)
````

````python
import catboost as cb

boosting_model = cb.CatBoostClassifier(
    n_estimators=200,
    logging_level='Silent',
    cat_features=cat_cols,
    eval_metric='AUC'
)

#   
param_grid = {
    'l2_leaf_reg': np.linspace(0, 1, 20)  #    
}

#  grid_search (CatBoost    -  split, 
#        X_train, y_train   ,  refit=True)
boosting_model.grid_search(
    param_grid=param_grid,
    X=X_train,
    y=y_train,
    refit=True,
    verbose=False
)
````

````python


y_train_pred = boosting_model.predict_proba(X_train)[:, 1]
y_test_pred = boosting_model.predict_proba(X_test)[:, 1]

train_auc = roc_auc_score(y_train, y_train_pred)
test_auc = roc_auc_score(y_test, y_test_pred)

print(f'train_auc_score: {train_auc}')
print(f'test_auc_score: {test_auc}')
````

````python
from sklearn.metrics import roc_curve
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_pred)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_pred)
````

````python
plt.figure(figsize=(8,6))
plt.plot(fpr_train, tpr_train, label=f'Train AUC={train_auc:.4f}')
plt.plot(fpr_test, tpr_test, label=f'Test AUC={test_auc:.4f}')
plt.plot([0,1], [0,1], linestyle='--', color='gray')
plt.legend(loc='lower right')
plt.show()
````
